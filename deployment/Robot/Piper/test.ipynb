```python
def eef_conversion_inv(pose):
    """
    Convert piper data(0.001mm and 0.001degree) to data suited policy(meter and radian)
    """
    # 将前三位从0.001毫米转换为米
    pose[:3] = [p / 1_000_000 for p in pose[:3]]
    # 将后三位从角度转换为弧度
    pose[3:6] = [np.deg2rad(p / 1000) for p in pose[3:6]]

    # TODO: 夹爪单位转换还未实现

    return np.asarray(pose)

# eef = [54868, 0, 260369, 0, 84929, 0]
eef = np.asarray([54868, 0, 260369, 0, 84929, 0], dtype=np.float64)
eef_ = eef_conversion_inv(eef)
print(np.asarray(eef_))
```
